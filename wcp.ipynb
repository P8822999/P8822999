```python
!pip install openpyxl
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
import xlsxwriter
```

```python
url = 'https://www.bot.or.th/english/statistics/_layouts/application/exchangerate/exchangerate.aspx'
ps = requests.get(url,'html.parser')
```

```python
soup = BeautifulSoup(ps.text)
```

```python
table1 = soup.find_all('table',class_='table')
```

```python
rows = []

for i in table1:
    a = i.find_all('tr',class_='bg-gray')
    for j in a:
        country = j.find_all('td',class_='tx-news')[0].text.strip()
        currency = j.find_all('td',class_='tx-news')[1].text.strip()
        buying_rate = j.find_all('td',class_='tx-news')[2].text.strip()
        selling_rate = j.find_all('td',class_='tx-news')[3].text.strip()
        rows.append((country,currency,buying_rate,selling_rate))
        
```

```python
df = pd.DataFrame(rows,columns=['Country','Currency','Buying rate','Selling rate'])
df['Buying rate'].astype(float)
df['Selling rate'].astype(float)
df.index = df.index + 1
df
```

```python
writer = pd.ExcelWriter('ExRate.xlsx')
df.to_excel(writer,index=False)
writer.save()
```
